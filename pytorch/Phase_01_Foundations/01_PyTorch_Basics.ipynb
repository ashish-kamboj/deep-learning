# PyTorch Basics - Tensors and Operations

## Objectives
- Understand PyTorch tensors and how they differ from NumPy arrays

- Master GPU vs CPU computation

## What is a PyTorch Tensor?
A tensor is the fundamental data structure in PyTorch - essentially a multi-dimensional array that can be computed on GPUs.

## What We're About to Do

The code below imports essential libraries. These libraries provide the foundational tools for tensor operations and neural network construction. Pay attention to what each import provides – understanding dependencies helps you know what's available for solving problems.


```python
# Import necessary libraries for tensor operations and deep learning
import torch
import numpy as np
import matplotlib.pyplot as plt

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```

## 1. Tensor Creation

```python
# Execute code with detailed step-by-step process
# Create tensor from Python list
x = torch.tensor([1, 2, 3])
print(f"From list: {x}")
print(f"Shape: {x.shape}")
print(f"Data type: {x.dtype}")

# Create 2D tensor
matrix = torch.tensor([[1, 2], [3, 4], [5, 6]])
print(f"\nMatrix:\n{matrix}")
print(f"Matrix shape: {matrix.shape}")
```

```python
# Execute code with detailed step-by-step process
# Common tensor creation methods
zeros = torch.zeros(3, 4)  # Matrix of zeros
ones = torch.ones(2, 3)    # Matrix of ones
random = torch.randn(2, 3) # Random normal distribution
uniform = torch.rand(2, 3) # Uniform [0, 1)

print(f"Zeros:\n{zeros}\n")
print(f"Ones:\n{ones}\n")
print(f"Random (normal):\n{random}\n")
print(f"Random (uniform):\n{uniform}")
```

```python
# Execute code with detailed step-by-step process
# Create tensor with specific range
arange = torch.arange(0, 10, 2)  # 0 to 10, step 2
linspace = torch.linspace(0, 1, 5)  # 5 points from 0 to 1
identity = torch.eye(3)  # Identity matrix

print(f"Arange: {arange}")
print(f"Linspace: {linspace}")
print(f"Identity:\n{identity}")
```

## 2. Tensor Properties

```python
# Execute code with detailed step-by-step process
x = torch.randn(2, 3, 4)

print(f"Shape: {x.shape}")
print(f"Size: {x.size()}")
print(f"Dimensions (ndim): {x.ndim}")
print(f"Number of elements: {x.numel()}")
print(f"Data type: {x.dtype}")
print(f"Device: {x.device}")
```

```python
# Execute code with detailed step-by-step process
# Different data types
int_tensor = torch.tensor([1, 2, 3], dtype=torch.int32)
float_tensor = torch.tensor([1.0, 2.0, 3.0], dtype=torch.float32)
double_tensor = torch.tensor([1.0, 2.0, 3.0], dtype=torch.float64)
bool_tensor = torch.tensor([True, False, True])

print(f"Int32: {int_tensor.dtype}")
print(f"Float32: {float_tensor.dtype}")
print(f"Float64: {double_tensor.dtype}")
print(f"Bool: {bool_tensor.dtype}")
```

## 3. Indexing and Slicing

```python
# Execute code with detailed step-by-step process
x = torch.arange(12).reshape(3, 4)
print(f"Tensor:\n{x}\n")

# Basic indexing
print(f"First row: {x[0]}")
print(f"Element [1,2]: {x[1, 2]}")
print(f"Last row: {x[-1]}")
```

```python
# Execute code with detailed step-by-step process
# Slicing
print(f"First two rows:\n{x[:2]}")
print(f"\nAll rows, first 2 columns:\n{x[:, :2]}")
print(f"\nEvery other element (rows):\n{x[::2]}")
print(f"\nEvery other element (cols):\n{x[:, ::2]}")
```

## 4. Reshaping and Transposing

```python
# Execute code with detailed step-by-step process
x = torch.arange(12)
print(f"Original: {x.shape} - {x}\n")

# Reshape
reshaped = x.reshape(3, 4)
print(f"Reshaped (3, 4):\n{reshaped}\n")

reshaped2 = x.reshape(2, 2, 3)
print(f"Reshaped (2, 2, 3):\n{reshaped2}\n")

# Flatten
flattened = reshaped.flatten()
print(f"Flattened: {flattened}")
```

```python
# Execute code with detailed step-by-step process
x = torch.arange(6).reshape(2, 3)
print(f"Original shape: {x.shape}")
print(f"Original:\n{x}\n")

# Transpose
transposed = x.T
print(f"Transposed shape: {transposed.shape}")
print(f"Transposed:\n{transposed}")
```

## 5. Mathematical Operations

```python
# Execute code with detailed step-by-step process
a = torch.tensor([1, 2, 3], dtype=torch.float32)
b = torch.tensor([4, 5, 6], dtype=torch.float32)

print(f"a: {a}")
print(f"b: {b}")
print(f"\nElement-wise add: {a + b}")
print(f"Element-wise sub: {a - b}")
print(f"Element-wise mul: {a * b}")
print(f"Element-wise div: {a / b}")
print(f"Power: {a ** 2}")
```

```python
# Execute code with detailed step-by-step process
# Statistical operations
x = torch.randn(2, 3, 4)
print(f"Mean: {x.mean()}")
print(f"Std: {x.std()}")
print(f"Sum: {x.sum()}")
print(f"Min: {x.min()}")
print(f"Max: {x.max()}")
print(f"Mean along axis 1 shape: {x.mean(dim=1).shape}")
```

```python
# Execute code with detailed step-by-step process
# Matrix multiplication
a = torch.randn(3, 4)
b = torch.randn(4, 2)

result = torch.matmul(a, b)
print(f"a shape: {a.shape}")
print(f"b shape: {b.shape}")
print(f"Result shape: {result.shape}")

# Using @ operator
result2 = a @ b
print(f"Same using @: {torch.allclose(result, result2)}")
```

## 6. Broadcasting

```python
# Execute code with detailed step-by-step process
# Broadcasting example
x = torch.randn(3, 1)
y = torch.randn(1, 4)

print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")
print(f"x + y shape: {(x + y).shape}")

# Explicit example
a = torch.tensor([[1], [2], [3]])  # shape (3, 1)
b = torch.tensor([10, 20, 30, 40])  # shape (4,)

result = a + b  # Broadcast to (3, 4)
print(f"\nBroadcasting example:\n{result}")
```

## 7. GPU vs CPU

```python
# Execute code with detailed step-by-step process
# Check CUDA availability
print(f"CUDA available: {torch.cuda.is_available()}")
print(f"GPU count: {torch.cuda.device_count()}")
if torch.cuda.is_available():
    print(f"Current GPU: {torch.cuda.get_device_name(0)}")
```

```python
# Create on CPU
x_cpu = torch.randn(3, 4)
print(f"x_cpu device: {x_cpu.device}")

# Move to GPU (if available)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
x_gpu = x_cpu.to(device)
print(f"x_gpu device: {x_gpu.device}")

# Alternative syntax
x_gpu2 = x_cpu.cuda() if torch.cuda.is_available() else x_cpu
print(f"x_gpu2 device: {x_gpu2.device}")
```

```python
# Best practice: use device variable
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create directly on device
x = torch.randn(3, 4, device=device)
y = torch.randn(3, 4, device=device)

# Operations on same device
z = x + y
print(f"Result device: {z.device}")
```

## 8. NumPy Interoperability

```python
# Execute code with detailed step-by-step process
# NumPy to PyTorch
np_array = np.array([[1, 2, 3], [4, 5, 6]])
torch_tensor = torch.from_numpy(np_array)

print(f"NumPy type: {type(np_array)}")
print(f"PyTorch type: {type(torch_tensor)}")
print(f"PyTorch dtype: {torch_tensor.dtype}")

# PyTorch to NumPy
torch_tensor2 = torch.randn(2, 3)
np_array2 = torch_tensor2.numpy()

print(f"\nNumPy shape: {np_array2.shape}")
print(f"NumPy dtype: {np_array2.dtype}")
```

## 9. In-place Operations

```python
# Execute code with detailed step-by-step process
x = torch.tensor([1.0, 2.0, 3.0])
print(f"Original x: {x}")
print(f"Original x id: {id(x)}")

# Out-of-place operation
y = x + 1
print(f"\nAfter y = x + 1:")
print(f"x: {x}")
print(f"y: {y}")
print(f"y id: {id(y)}")

# In-place operation (note the underscore)
x.add_(1)
print(f"\nAfter x.add_(1):")
print(f"x: {x}")
print(f"x id: {id(x)}")
```

## 10. Practice Exercise

**Task**: Create a batch of 32 random images (3 channels, 224x224 pixels each), move to GPU, compute mean and std

```python
# Execute code with detailed step-by-step process
# Solution
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create batch of images
batch_size = 32
channels = 3
height, width = 224, 224

images = torch.randn(batch_size, channels, height, width, device=device)

print(f"Batch shape: {images.shape}")
print(f"Device: {images.device}")
print(f"Mean: {images.mean():.4f}")
print(f"Std: {images.std():.4f}")
```
