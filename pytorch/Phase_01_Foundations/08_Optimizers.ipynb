```python
# Define a custom function with detailed implementation
## 6. Compare Optimizers on Simple Problem

def train_with_optimizer(optimizer_class, optimizer_kwargs, epochs=200, lr=0.01):
    """Train a simple model with given optimizer"""
    model = nn.Sequential(
        nn.Linear(2, 64),
        nn.ReLU(),
        nn.Linear(64, 1)
    )
    
# Update model parameters based on computed gradients
    optimizer = optimizer_class(model.parameters(), lr=lr, **optimizer_kwargs)
# Compute the loss (error) between predictions and actual values
    criterion = nn.MSELoss()
    
    # Simple dataset
    X = torch.randn(100, 2)
    y = (X[:, 0] + X[:, 1] > 0).float().unsqueeze(1)
    
# Compute the loss (error) between predictions and actual values
    losses = []
    
    model.train()
# Iterate through batches of data
    for epoch in range(epochs):
        output = model(X)
# Compute the loss (error) between predictions and actual values
        loss = criterion(output, y)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        losses.append(loss.item())
    
    return losses

# Train with different optimizers
results = {}
results['SGD'] = train_with_optimizer(optim.SGD, {}, epochs=200, lr=0.01)
results['SGD+Momentum'] = train_with_optimizer(optim.SGD, {'momentum': 0.9}, epochs=200, lr=0.01)
results['Adam'] = train_with_optimizer(optim.Adam, {}, epochs=200, lr=0.001)
results['RMSprop'] = train_with_optimizer(optim.RMSprop, {}, epochs=200, lr=0.001)
results['AdamW'] = train_with_optimizer(optim.AdamW, {}, epochs=200, lr=0.001)

print("Training completed.")
```

## Defining the Loss Function

The loss function measures how wrong our predictions are. During training, we'll minimize this value. Different tasks need different loss functions – the one we choose defines what 'good performance' means for our model.


```python
# Configure loss function and optimization algorithm
## 7. Visualize Optimizer Convergence

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Full training curve
# Iterate through batches of data
for name, losses in results.items():
# Compute the loss (error) between predictions and actual values
    axes[0].plot(losses, label=name, linewidth=2)

axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Optimizer Convergence')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_yscale('log')

# Last 50 epochs zoomed
# Iterate through batches of data
for name, losses in results.items():
# Compute the loss (error) between predictions and actual values
    axes[1].plot(losses[-50:], label=name, linewidth=2)

axes[1].set_xlabel('Epoch (last 50)')
axes[1].set_ylabel('Loss')
axes[1].set_title('Final Convergence (Zoomed)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print final losses
print("\nFinal Loss After 200 Epochs:")
print("="*40)
# Iterate through batches of data
for name, losses in results.items():
    print(f"{name:15s}: {losses[-1]:.6f}")
```

## Building the Model

```python
# Set up the neural network model architecture
## 10. Gradient Clipping and Optimizer State

model = nn.Linear(10, 1)
# Update model parameters based on computed gradients
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training with gradient clipping
X = torch.randn(100, 10)
y = torch.randn(100, 1)

# Compute the loss (error) between predictions and actual values
criterion = nn.MSELoss()

# Iterate through batches of data
for step in range(5):
    output = model(X)
# Compute the loss (error) between predictions and actual values
    loss = criterion(output, y)
    
    optimizer.zero_grad()
    loss.backward()
    
    # Gradient clipping
# Update model parameters based on computed gradients
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    
    optimizer.step()
    
# Compute the loss (error) between predictions and actual values
    print(f"Step {step+1}: Loss = {loss.item():.4f}")

# Inspect optimizer state
print("\nOptimizer State:")
print(f"Number of parameter groups: {len(optimizer.param_groups)}")
# Iterate through batches of data
for key, value in optimizer.param_groups[0].items():
# Update model parameters based on computed gradients
    if key != 'params':
        print(f"{key}: {value}")
```
