## Building the Model

Now we'll define our neural network architecture. Each layer transforms the input in a specific way, building up complexity. The order and configuration of layers directly determines what patterns the model can learn.


```python
# Define a custom function with detailed implementation
## Transformer Architecture
class TransformerEncoder(nn.Module):
    def __init__(self, d_model=512, num_layers=6, num_heads=8, d_ff=2048):
        super().__init__()
        self.embedding = nn.Embedding(1000, d_model)
        self.pos_encoding = self._get_pos_encoding(d_model, 512)
        self.layers = nn.ModuleList([
            nn.TransformerEncoderLayer(d_model, num_heads, d_ff, batch_first=True)
# Iterate through batches of data
            for _ in range(num_layers)
        ])
    
    def _get_pos_encoding(self, d_model, max_len):
        pe = torch.zeros(max_len, d_model)
        pos = torch.arange(0, max_len).unsqueeze(1).float()
        div = torch.exp(torch.arange(0, d_model, 2).float() * -(np.log(10000)/d_model))
        pe[:, 0::2] = torch.sin(pos * div)
        pe[:, 1::2] = torch.cos(pos * div)
        return pe.unsqueeze(0)
    
    def forward(self, x):
        x = self.embedding(x)
# Iterate through batches of data
        for layer in self.layers:
            x = layer(x)
        return x

model = TransformerEncoder()
print("Transformer: 6 layers × 8 heads + feed-forward")
print("Advantages: Parallel processing, long-range dependencies, strong NLP")
```

```python
# Define a custom function with detailed implementation
## Classification with BERT
class BERTClassifier(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        self.bert = BertModel.from_pretrained('bert-base-uncased')
        self.classifier = nn.Sequential(
            nn.Dropout(0.1),
            nn.Linear(768, 256),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(256, num_classes)
        )
    
    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids, attention_mask)
        cls_output = outputs.last_hidden_state[:, 0]  # [CLS] token
        logits = self.classifier(cls_output)
        return logits

print("BERT Fine-tuning Strategy:")
print("1. Use pretrained BERT weights")
print("2. Add task-specific head")
print("3. Train with low learning rate (2e-5)")
print("4. Freeze early layers (optional)")
print("5. Full fine-tuning often works best")
```
